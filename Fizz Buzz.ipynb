This coding chalenge is for the maths game "Fizz Buzz" where each player counts inturn from 1 to say 100 but:
- If the number is divisible by **3** the the player must say **"FIZZ!"** instead of the number.
- The same occurs if the number is divisible by **5** where the player must say **"BUZZ!'**.
- If the number is divisible by **both 3 AND 5** then the player must say **"FIZZ BUZZ!"**.

The key to this coding chalenge is not only to test the coding logic but also to be able to think about adaptability and scalability: 
- *What happens if the numbers 3 & 5 are changed to different values?*
- *What happens if more numbers need to be added? How easy will you code be to edit afterwards?*

Will your code need to have multiple values altered throughout the script before it can be updated to run with the new input values?

Below are two versions of the same problem:

```python
for i in range(1,101):
    if i%3 == 0 and i%5 == 0:
        print('FizzBuzz')
    elif i%3 == 0:
        print('Fizz')
    elif i%5 == 0:
        print('Buzz')
    else:
        print(i)
```

Although the version above works, we've hard coded the values into the script which is not ideal. 
To now change the values, words or even add in more values, we have to do extrensive re-working of the script to make the appropriate changes.
This version is neither adaptable nor scalable.

The alternative (badass) version is shown below:

```python
def fizzBuzz(inputValues, *args):
    for i in range(*args):
        output = ''
        for value in inputValues:
            if i%value == 0:
                output += inputValues[value]
        if output == '':
            output = i
        print(output)
        
inputValues = {3: 'Fizz', 5: 'Buzz'}
fizzBuzz(inputValues, 1,101,1)
```

Variable inputValues has been set up as a dictionary with key value pairs for 3 (Fizz) and 5 (Buzz). The advantage of this is that we can add or edit the dictionary easily and this data will be passed into the fizzBuzz function as a parameter. The arbitrary argument *args has been used so that the function can receive a tuple of arguments that can be adjusted as required - very cool! The example above shows the fizzBuzz() function being passed the dictionary pairs from variable inputValues and we can control remotely how the range() function works too; all from one place.

```python
def fizzBuzz(inputValues, *args):
    for i in range(*args):
        output = ''
        for value in inputValues:
            if i%value == 0:
                output += inputValues[value]
        if output == '':
            output = i
        print(output)
        
inputValues = {3: 'Fizz', 5: 'Buzz', 7: 'Pop', 17: 'Burp', 29: 'Belch'}
fizzBuzz(inputValues, 1,151,1)
```

We've very quickly refactored the code to accept additional parameters and increased the range to 150 by just editing two lines of code - this version is very scalable.
